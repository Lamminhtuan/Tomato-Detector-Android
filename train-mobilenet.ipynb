```python
import os
import json
import tensorflow as tf
from mediapipe_model_maker import object_detector
```

```python
dataset_folder = "/kaggle/working/dataset"
train_image_folder = f"{dataset_folder}/train/images"
val_image_folder = f"{dataset_folder}/val/images"
train_label_folder = f"{dataset_folder}/train/Annotations"
val_label_folder = f"{dataset_folder}/val/Annotations"
```

```python
import shutil
```

```python
train_path = "/kaggle/input/augmented-tomato-last/dataset/train"
val_path = "/kaggle/input/augmented-tomato-last/dataset/val"
```

```python
train_data = object_detector.Dataset.from_pascal_voc_folder(train_path, cache_dir = "/tmp/data/train")
val_data = object_detector.Dataset.from_pascal_voc_folder(val_path, cache_dir = "/tmp/data/val")
print(f"Training size: {train_data.size}")
print(f"Validation size: {val_data.size}")
```

```python
spec = object_detector.SupportedModels.MOBILENET_V2_I320
hparams = object_detector.HParams(export_dir = '/kaggle/working/exported_model')
hparams.epochs = 50
hparams.shuffle = True
hparams.batch_size=16
options = object_detector.ObjectDetectorOptions(
    supported_model=spec,
    hparams=hparams
)
```

```python
model = object_detector.ObjectDetector.create(
    train_data=train_data,
    validation_data=val_data,
    options=options
)
```

```python
loss, metrics = model.evaluate(val_data, batch_size=8)
print(loss)
print(metrics)
```
